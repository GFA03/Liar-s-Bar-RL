```python
from env import LiarsBarEnv
from qagent import QLearningAgent

env = LiarsBarEnv(num_players=4)
agent = QLearningAgent(env)

done = False
observation, info = env.reset()
env.action_space.sample()
```

```python
num_episodes = 100
for i in range(num_episodes):
    done = False
    while not done:
        env.render()
        action = agent.get_action(observation)
        # If there is only one action available, take it
        if (len(env._get_available_actions()) == 1):
            action = env._get_available_actions()[0]
        new_observation, reward, done, info = env.step(action)
        agent.learn(observation, action, reward, new_observation, done)
        observation = new_observation
    observation, info = env.reset()
    print("Episode {} done".format(i))

print("Training done")
```
